```python
import pandas as pd
import json
import numpy
from splinter import Browser
from bs4 import BeautifulSoup
import requests
```

```python
golf_data = pd.read_csv("data (version 1).csv")
golf_data
```

```python

golf_data = golf_data.loc[:, ["Name", "Driving Distance", "Driving Accuracy Percentage", "Greens in Regulation Percentage", 
                             "Putts Per Round", "Rough Tendency", "Scrambling", "Proximity to Hole", "Approaches from inside 100 yards", 
                             "Approaches from > 100 yards", "Sand Save Percentage"]]
golf_data
```

```python
url = 'https://www.espn.com/golf/rankings'

# Use pandas to read the table on the page
dfs = pd.read_html(url)

# Select the first table, which contains the world golf rankings
world_rankings = dfs[0]

# Display the rankings
world_rankings = world_rankings.rename(columns={'NAME': 'Name'})
world_rankings
```

```python
merged_df = pd.merge(golf_data, world_rankings, on='Name')
merged_df
```

```python
merged_df = merged_df.dropna()
merged_df
```

```python
top_10 = merged_df[(merged_df['RK'] >= 1) & (merged_df['RK'] <= 10)]
averages_top_10 = top_10.mean()
averages_top_10
```

```python
next_11_25 = merged_df[(merged_df['RK'] >= 11) & (merged_df['RK'] <= 25)]
averages_11_25 = next_11_25.mean()
averages_11_25
```

```python
next_26_50 = merged_df[(merged_df['RK'] >= 26) & (merged_df['RK'] <= 50)]
averages_26_50 = next_26_50.mean()
averages_26_50
```

```python
next_51_100 = merged_df[(merged_df['RK'] >= 51) & (merged_df['RK'] <= 100)]
averages_51_100 = next_51_100.mean()
averages_51_100
```

```python
over_100 = merged_df[(merged_df['RK'] >= 101)]
averages_over_100 = over_100.mean()
averages_over_100
```

```python
all_averages = pd.concat([averages_top_10, averages_11_25, averages_26_50, averages_51_100, averages_over_100], axis = 1)
all_averages.columns = ["Top Ten", "11-25", "26-50", "51-100", "Over 100"]
all_averages
```
